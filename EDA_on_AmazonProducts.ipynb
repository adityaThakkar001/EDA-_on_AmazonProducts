EDA on Amazon Products

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Amazon_Products.csv',encoding = 'unicode_escape')
```

```python
unnamed_columns = [col for col in df.columns if 'Unnamed' in col]
df.drop(columns=unnamed_columns, inplace=True)
```

```python
df
```

```python
df.dropna(inplace = True)
```

```python
df
```

```python
df['price'] = df['price'].str.replace('Â', '')
df['price'] = df['price'].str.replace('£', '')
df['number_available_in_stock'] = df['number_available_in_stock'].str.replace('Â', '')
df['number_available_in_stock'] = df['number_available_in_stock'].str.replace('new', '')
```

```python
df
```

```python
are_duplicates = df.duplicated(subset=['product_description', 'description'])
print("Duplicates present:", any(are_duplicates))
```

```python
df = df.drop(columns=['description'])
df
```

```python
df.columns
```

```python
split_column = df['amazon_category_and_sub_category'].str.split('>')
df['product_category'] = split_column.str[0]
df['product_subcategory'] = split_column.str[1:] 
```

```python
df
```

```python
df = df.drop(columns=['amazon_category_and_sub_category'])
df
```

```python
pal = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#9467bd"]
popular_categories = (df["product_category"].value_counts(ascending = True)).plot(
    kind="barh", color=pal, width=1, edgecolor="black"
)
plt.title("Most Popular Product Categories on Amazon",fontsize = 20)
plt.xlabel("product categories",fontsize = 15)
plt.yticks(fontsize = 12)
plt.xticks(fontsize = 15)
plt.figure(figsize=(20,20))
plt.show()
```

Mean and Median Price of Amazon Products for different Product Categories

```python
df["price"] = pd.to_numeric(df["price"], errors="coerce")
df["price"] = df["price"].round(2)
category_stats = df.groupby("product_category")["price"].agg(["mean", "median"])
category_stats
```

Mean and Median Number of Reviews of Amazon Products for different Product Categories

```python
df["number_of_reviews"] = pd.to_numeric(df["number_of_reviews"], errors="coerce")
df["number_of_reviews"] = df["number_of_reviews"].round(2)
review_stats = df.groupby("product_category")["number_of_reviews"].agg(["mean", "median"])
review_stats
```

Mean and Median of Review Ratings of Amazon Products for different Product Categories

```python
df['average_review_rating'] = df['average_review_rating'].str.replace(' out of 5 stars', '')
```

```python
df["average_review_rating"] = pd.to_numeric(df["average_review_rating"], errors="coerce")
df["average_review_rating"] = df["average_review_rating"].round(2)
rate_stats = df.groupby("product_category")["average_review_rating"].agg(["mean", "median"])
rate_stats
```

Price Distribution of Products across different Product Categories

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.histplot(x="product_category", y="price", data=df)
plt.xticks(rotation=90)  
plt.xlabel("Product Category")
plt.ylabel("Price")
plt.title("Price Distribution of Products by Category")
plt.tight_layout()
plt.show()
```

```python
sns.set(style="whitegrid")
custom_palette = sns.color_palette("Set1", n_colors=len(df["product_category"].unique()))
popular_categories = df["product_category"].value_counts(ascending=True)
category_popularity_mapping = {category: popularity for category, popularity in zip(popular_categories.index, popular_categories)}
df["popularity"] = df["product_category"].map(category_popularity_mapping)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="average_review_rating", y="popularity", hue="product_category")
plt.xlabel("Average Review Rating")
plt.ylabel("Popularity")
plt.title("Relation between Average Review Rating and Popularity by Product Category")
plt.legend(bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```
